### Exercise 1 — Even or Odd
Create a variable `number = 7`. Print `"Even"` if it is even, otherwise `"Odd"`.
*Hint: a number is even if `number % 2 == 0`.*

```python
# Your code here
number = 7

if number % 2 == 0 :
    print("Even number")
else:
    print("Odd number")
```

### Exercise 2 — Temperature check
Given `temp = 38`, print:
- `"Hot"` if temp is above 30
- `"Warm"` if temp is between 20 and 30 (inclusive)
- `"Cold"` otherwise

```python
# Your code here
temp = 38

if temp > 30:
    print("Hot")

elif temp > 20 and temp <= 30:
    print("Warm")

else:
    print("Cold")
```

### Exercise 3 — Login check
Given `username = "admin"` and `password = "1234"`, print `"Access granted"` only if BOTH the username is `"admin"` AND the password is `"1234"`. Otherwise print `"Access denied"`.

```python
# Your code here
username = input("Enter username")
password = input("Enter password")

if username == "admin" and password == "1234":
    print("Access granted")

else:
    print("Access denied")
```

### Exercise 6 — Traffic light 🚦
Given `light = "yellow"`, print:
- `"Stop"` if red
- `"Go"` if green
- `"Slow down"` if yellow
- `"Invalid light"` for anything else

```python
# Your code here
light = "yellow"

if light == "red":
    print("Stop")
elif light == "yellow":
    print("Slow down")
elif light == "green":
    print("Go")
else:
    print("Invalid light")
```
